```python
def get_heatmap(cuisine):
    cuisine_df = df[df['cuisines'].str.contains(cuisine)]
    cuisine_df_rest = cuisine_df['location'].value_counts().reset_index()
    cuisine_df_rest.columns = ['name', 'count']#.rename(columns={'index':'name',"location":"count"})
    heatmap_df = cuisine_df_rest.merge(locationss, on='name',how='left')
    print(heatmap_df.head())
    basemap = folium.Map()
    HeatMap(heatmap_df[['latitude','longitude','count']]).add_to(basemap)
    return basemap
```
